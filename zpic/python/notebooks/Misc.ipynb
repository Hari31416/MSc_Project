```python
import sys
sys.path.append("../lib")
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,8)
```

## Moving Window

```python
import em1d as zpic

dt = 0.019

nx  = 1000

box = 20.0

sim = zpic.Simulation( nx, box, dt )

sim.add_laser( zpic.Laser( start = 15.0, fwhm = 4.0, a0 = 1.0, omega0 = 10.0,
               polarization = np.pi/2 ))


iterations = int(50/dt)
Ezs = np.zeros((27, nx))
Bys = np.zeros((27, nx))
for i in range(iterations):
    print(f"{i+1} of {iterations}", end="\r")
    sim.iter()
    if i%100 == 0:
        Ezs[i//100] = sim.emf.Ez
        Bys[i//100] = sim.emf.By
```

```python
def plot_field(iter):
    xmin = sim.dx/2 + sim.t
    xmax = sim.box + (sim.n_move - 0.5) * sim.dx
    time = 100*dt*iter
    plt.plot(np.linspace(xmin, xmax, num = sim.nx), Ezs[iter], label = "$E_z$ field" )
    plt.plot(np.linspace(xmin, xmax, num = sim.nx), Bys[iter], label = "$B_y$ field" )
    plt.xlabel("x")
    plt.ylabel("Field")
    plt.title(f"Time = {time}")
    plt.legend()
    plt.show()
```

```python
import em1d as zpic

dt = 0.019

nx  = 1000

box = 20.0

sim = zpic.Simulation( nx, box, dt )

sim.add_laser( zpic.Laser( start = 15.0, fwhm = 4.0, a0 = 1.0, omega0 = 10.0,
               polarization = np.pi/2 ))


iterations = int(50/dt)
Ezs = []
Bys = []
for i in range(iterations):
    print(f"{i+1} of {iterations}", end="\r")
    sim.iter()
    if i%100 == 0:
        Ezs.append(sim.emf.Ez)
        Bys.append(sim.emf.By)
```

```python
for i in range(0, 30, 5):
    plot_field(i)
```

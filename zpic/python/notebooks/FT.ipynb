```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
plt.style.use('ggplot')
```

```python
f1 = 3.0
f2 = 5.0
t = np.linspace(0, 1, 1000)
x1 = np.sin(2 * np.pi * f1 * t)
x2 = np.sin(2 * np.pi * f2 * t)
plt.figure(figsize=(12,8))
plt.plot(t, x1, label='$x_1(t)$')
plt.plot(t, x2, label='$x_2(t)$')
plt.plot(t, x1+x2, label='$x_1(t)+x_2(t)$')
plt.legend(loc='upper right')
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.show()
```

```python
x1_f = np.fft.fft(x1)
x2_f = np.fft.fft(x2)
x_f_p = x1_f + x2_f
x_f = np.fft.fft(x2+x1)
```

```python
plt.figure(figsize=(12,8))
plt.plot(np.abs(x1_f), label='$x_1(t)$')
plt.plot(np.abs(x2_f), label='$x_2(t)$')
plt.plot(np.abs(x_f_p), label='$x_1(t)+x_2(t)$')
plt.xlim(0, 10)
plt.legend(loc='upper right')
```

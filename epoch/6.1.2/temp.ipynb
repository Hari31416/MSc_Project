```python
import sdf_helper as sh
import matplotlib.pyplot as plt
import os
from matplotlib import colors
import sdf
```

```python
data.Electric_Field_Ey.data.max()
```

```python
def plot_field(data, ax, t):
    data = sdf.read(data)
    variable = data.Electric_Field_Ey.data
    variable = variable/variable.max()
    # print(variable.max(), variable.min())
    ax.imshow(variable**2, cmap='gray', origin='lower',
    extent=[-10,10, -25, 25],
    aspect='auto',
    interpolation='nearest',
    norm=colors.Normalize(vmin=-0.05, vmax=0.05)
    )
    ax.set_xlabel('$y \, [\mu m]$')
    ax.set_ylabel('$x \, [\mu m]$')
    ax.set_title(f't = {t} fs')
```

```python
plot_field('0010.sdf', plt.gca(), 10)
```

```python
all_files = [f for f in os.listdir('.') if f.endswith('.sdf')]
```

```python
fig, ax = plt.subplots(4, 5, figsize=(25, 22))
i=0
for data in all_files[1:]:
    t = (i+1)*10
    plot_field(data, ax[i // 5, i % 5], t)
    i += 1
```

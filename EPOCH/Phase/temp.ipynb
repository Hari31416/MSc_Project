```python
import sdf_helper as sh
import matplotlib.pyplot as plt
import os
import sdf
```

```python
data = sh.getdata('0010.sdf')
```

```python
def plot_field(data, ax):
    data = sdf.read(data)
    variable = data.Electric_Field_Ey.data
    ax.imshow(variable**2, cmap='gray', origin='lower')
```

```python
all_files = [f for f in os.listdir('.') if f.endswith('.sdf')]
```

```python
fig, ax = plt.subplots(3, 5, figsize=(15, 10))
i=0
for data in all_files[:-1]:
    plot_field(data, ax[i // 5, i % 5])
    i += 1
```

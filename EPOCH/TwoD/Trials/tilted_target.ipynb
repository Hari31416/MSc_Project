```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def box_function(x, length):
    """Box function of length centered at x"""
    return (np.abs(x) < length / 2).astype(float)

X = np.linspace(-1, 1, 1000)
Y = box_function(X, 0.2)
plt.plot(X, Y)
```

```python
theta = np.pi/3
x0 = 0
y0 = 7
thickness = 1
length = 4*np.cos(theta)

def inclided_target(x, theta, x0, y0, thickness, length):
    m = - np.tan(theta)
    if np.abs(x) <= length / 2:
        mask = 1
    else:
        mask = 0
    x = mask * x
    y = m*(x-x0) + y0 - thickness
    return y*mask
```

```python
X = np.linspace(-10, 10, 100)

Y1 = [inclided_target(x, theta, x0, y0, thickness=0, length=length) for x in X]
Y2 = [inclided_target(x, theta, x0, y0, thickness=thickness, length=length) for x in X]
plt.plot(X, Y1, 'r')
plt.plot(X, Y2, 'b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(y0-length/2, color='k')
```

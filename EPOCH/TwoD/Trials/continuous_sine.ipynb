```python
def x_profile(t, omega, Tau):
    T = 2 * np.pi / omega
    if T > Tau:
        return np.sin(omega * t)
    return 0
def y_profile(t, omega, Tau):
    T = 2 * np.pi / omega
    Omega = 2 * np.pi / Tau
    if T > Tau:
        return np.sin(Omega * t)
    return 0

def time_profile(t, omega, Tau):
    T = 2 * np.pi / omega
    Omega = 2 * np.pi / Tau
    if T > Tau:
        return np.sin(Omega * t)
    return 0

def profile(t, omega, Tau):
    T = 2 * np.pi / omega
    Omega = 2 * np.pi / Tau
    if t > Tau:
        return np.sin(Omega * t)
    return 0
```

```python
time_period = 1
omega = 2 * np.pi / time_period
Tau = 5*time_period
ts = np.linspace(0, 10*time_period, 100)
xs = np.linspace(0, 10, 100)
ys = np.linspace(0, 10, 100)
x_prof = [profile(x, omega, Tau) for x in xs]
y_prof = [profile(y, omega, Tau) for y in ys]
t_prof = [profile(t, omega, Tau) for t in ts]
```

```python
plt.plot(xs, x_prof)
plt.plot(ys, y_prof)
plt.plot(ts, t_prof)
```

```python
omegax = 1
omegay = 1
kx = 1
ky = 1

def sin_curve(x, y, t):
    return np.sin(kx*x - omegax*t)*np.sin(ky*y - omegay*t)
```

```python
X = np.linspace(-np.pi, np.pi, 100)
Y = np.linspace(-np.pi, np.pi, 100)
t = np.linspace(0, 10, 100)

Xs, Ys = np.meshgrid(X, Y)
```

```python
from IPython.display import clear_output
import time

for i in range(len(t)):
    clear_output(wait=True)
    plt.imshow(sin_curve(Xs, Ys[0], t[i]))
    plt.show()
    time.sleep(0.1)
```

```python
sines = sin_curve(Xs, Ys, t[16])
plt.contourf(Xs, Ys, sines)
plt.colorbar()
```

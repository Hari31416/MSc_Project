```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_DIR = "output_data"
XLIM = (0, 20)
YLIM = (1e-3, 1.2)
```

```python
def get_ticks(type_):
    if type_ in ["o", "odd"]:
        return np.arange(1, 21, 2)
    elif type_ in ["e", "even"]:
        return np.arange(2, 21, 2)
    elif type_ in ["a", "all", "b", "both"]:
        return np.arange(1, 21, 1)
    else:
        return None
```

```python
def plot_one_fft(ax, run, id_, component, xlim = XLIM, ylim = YLIM, hline = True, ticks="odd"):
    E_file_name = f"{DATA_DIR}/{run}_fft_{component}_{id_}.npy"
    omega_file_name = f"{DATA_DIR}/{run}_omega_{component}_{id_}.npy"
    E = np.load(E_file_name)
    omega = np.load(omega_file_name)
    ax.plot(omega, E, label = f"$E_{component}$")
    ax.set_xlabel("omega")
    ax.set_ylabel("E")
    ax.set_title(f"{run} {id_}")
    ax.set_yscale("log")

    xticks = get_ticks(ticks)
    ax.set_xticks(xticks)
    if hline:
        for xtick in xticks:
            ax.axvline(xtick, c="k", linestyle = "--", alpha = 0.5)
    if xlim is not None:
        ax.set_xlim(*xlim)
    if ylim is not None:
        ax.set_ylim(*ylim)
    # plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (8, 5))
plot_one_fft(ax, "1run", "1", "x", ticks="both")
plot_one_fft(ax, "1run", "1", "y", ticks="both", hline=False)
ax.legend();
```

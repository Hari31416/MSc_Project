```python
import sdf
import matplotlib.pyplot as plt
import numpy as np

#increase plt fint sizes
plt.rcParams['font.size']=20
plt.rcParams['legend.fontsize']=20
```

```python
ppt_path = "/media/hari31416/Hari_SSD/Users/harik/Desktop/MSc_Project/Presentations/PPTs/Sem_4_Major/images"
```

```python
m = 9.10938356e-31
e = 1.60217662e-19
c = 299792458
PI = np.pi
epsilon = 8.85e-12
kb = 1.38064852e-23
LAMBD = 1e-6
omega0 = 2*PI*c/LAMBD
nc = epsilon*m*omega0**2/e**2
```

```python
raw_data = sdf.read('1run/0000.sdf',dict=True)
density = raw_data['Derived/Number_Density/Electron'].data
```

```python
density = density/nc
```

```python
EXTENT = [-8, 2, -10, 10]
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
im = ax.imshow(density.T, cmap='jet', extent=EXTENT, origin='lower', aspect='auto')
ax.set_xlabel(r'$x\ [\lambda]$')
ax.set_ylabel(r'$y\ [\lambda]$')
ax.set_title('Simulation Box')
# fig.colorbar(im, ax=ax)
#set colorbar label
cbar = fig.colorbar(im, ax=ax)
cbar.set_label(r'$n_e\ [n_c]$')
fig.tight_layout()
fig.savefig("images_f/2dbox.png", bbox_inches='tight')
fig.savefig(ppt_path+"/2dbox.png", bbox_inches='tight')
```

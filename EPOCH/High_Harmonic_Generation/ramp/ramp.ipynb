```python
import numpy as np
import matplotlib.pyplot as plt
```

# The Function

```python
max_density = 8
length = 1
ramp_length = 10
ramp_start = 10
def ramp_density(x):
    if x < ramp_start:
        return 0
    elif x > ramp_start + ramp_length:
        return 0
    else:
        return (x - ramp_start) / ramp_length
    
step_start = ramp_start + ramp_length
step_length = 1
def step_density(x):
    if x < step_start:
        return 0
    elif x > step_start + step_length:
        return 0
    else:
        return 1
```

```python
x = np.linspace(0, 40, 1000)
r_d = np.array([ramp_density(i) for i in x])
s_d = np.array([step_density(i) for i in x])
d = (r_d+s_d)*max_density
plt.plot(x, d)
```

```python
raw_data = sdf.read(all_sdfs[5])
```

```python
m = 9.10938356e-31
e = 1.60217662e-19
c = 299792458
PI = np.pi
epsilon = 8.85e-12
lambd = 1e-6
omega0 = 2 * PI * c / lambd
nc = epsilon*m*omega0**2/e**2
```

```python
d = raw_data.Derived_Number_Density_Electron.data
x = raw_data.Grid_Grid_mid.data[0]
plt.plot(x/lambd, d/nc)
```

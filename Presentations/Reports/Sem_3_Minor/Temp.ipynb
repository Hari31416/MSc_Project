```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.size"]=14
```

```python
def sine(A, f, phi, t):
    return A * np.sin(2*np.pi*f * t + phi)
```

```python
def sine(A, f, phi, t):
    return A * np.sin(2*np.pi*f * t + phi)

t = np.linspace(0, 2 * np.pi, 100)
x1 = sine(0.1, 5.5, 0.1, t)
x2 = sine(0.2, 6.1, 0.0, t)
x3 = sine(0.2, 4.9, 0.45, t)
x = x1+x2+x3
plt.figure(figsize=(10, 8))
plt.plot(t, x)
plt.title("Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xticks([])
plt.yticks([])
plt.show()

plt.figure(figsize=(10, 8))
y = np.fft.fft(x)
plt.plot(np.abs(y)[len(y)//2:])
plt.xticks([])
plt.yticks([])
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.title("Fourier Transform")
plt.show()
```

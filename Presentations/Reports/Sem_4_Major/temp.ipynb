```python
import matplotlib.pyplot as plt

# Define the figure and axes
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
fig.suptitle('Particle-in-Cell (PIC) Cycle')

# Initialize the simulation
axs[0, 0].set_title('1. Initialize')
axs[0, 0].axis('off')
axs[0, 0].annotate('Initial positions, velocities, and charges\nElectric and magnetic fields set to zero', (0.5, 0.5),
                    xycoords='axes fraction', ha='center', va='center')

# Electromagnetic field calculation
axs[0, 1].set_title('2. Electromagnetic\nField Calculation')
axs[0, 1].axis('off')
axs[0, 1].annotate('Electric and magnetic fields calculated\nfrom charge and current densities', (0.5, 0.5),
                    xycoords='axes fraction', ha='center', va='center')

# Particle motion calculation
axs[0, 2].set_title('3. Particle Motion\nCalculation')
axs[0, 2].axis('off')
axs[0, 2].annotate('Particle positions and velocities\nupdated using electromagnetic forces', (0.5, 0.5),
                    xycoords='axes fraction', ha='center', va='center')

# Charge density and current density calculation
axs[1, 0].set_title('4. Charge Density\nand Current Density\nCalculation')
axs[1, 0].axis('off')
axs[1, 0].annotate('Charge density and current density\nupdated using new particle positions and velocities', (0.5, 0.5),
                    xycoords='axes fraction', ha='center', va='center')

# Re-calculation of electromagnetic fields
axs[1, 1].set_title('5. Re-calculation\nof Electromagnetic Fields')
axs[1, 1].axis('off')
axs[1, 1].annotate('Electric and magnetic fields re-calculated\nusing updated charge and current densities', (0.5, 0.5),
                    xycoords='axes fraction', ha='center', va='center')

# Repeat
axs[1, 2].set_title('6. Repeat')
axs[1, 2].axis('off')
axs[1, 2].annotate('Cycle repeated until simulation time is reached', (0.5, 0.5),
                    xycoords='axes fraction', ha='center', va='center')

plt.show()
```
